```python
for index, row in data.iterrows():
    sexo_code = row['SEXO']
    # print(sexo_code)
    reg_code = row['REG_POND']
    sexo_label = lista_labels.loc[(lista_labels['Coluna'] == 'SEXO') & (lista_labels['Codigo'] == sexo_code), 'Label'].values[0]
    reg_label = lista_labels.loc[(lista_labels['Coluna'] == 'REG_POND') & (lista_labels['Codigo'] == reg_code), 'Label'].values[0]
    combined_label = f"{sexo_label} - {reg_label}"
    data.at[index, 'SEXO_REG_label'] = combined_label
data[['SEXO', 'REG_POND', 'SEXO_REG', 'SEXO_REG_label']]
```

```python
lista_labels = pd.concat([lista_labels, pd.DataFrame(dict_teste)], axis=0, ignore_index=True)
display(lista_labels)
```

## <font color="greenblue">Lógica para fazer um Recode</font>

```python
import pandas as pd
import numpy as np
data_file = r'C:\Users\rayner.santos\Downloads\Consistencia e Processamento\BD_CNP_Recall de campanha_v12.xlsx'
data = pd.read_excel(data_file, sheet_name='Códigos')
lista_labels = pd.read_excel(data_file, sheet_name='Variáveis Labels')
lista_labels = lista_labels.iloc[1:, :].copy()
lista_labels.columns = ['Coluna', 'Codigo', 'Label']
lista_labels["Coluna"] = lista_labels["Coluna"].ffill().str.strip()
display(data)
```

```python
display(lista_labels)
```

<font color="gray50">Pensando que desejo fazer um recode para a variável Q7 (Estados)</font>

```python
NOME_COLUNA_RECODE = 'Q7'
NOME_BANDEIRA_RECODE = 'BANDEIRA_NOVA'
lista_labels.loc[lista_labels['Coluna'] == NOME_COLUNA_RECODE]
```

<font color="gray33">Regras dada pelo usuário</font>

```python
BANDEIRA = {
    1: 'São Paulo',
    2: 'Minas Gerais',
    3: 'Rio de Janeiro',
    4: 'Espírito Santo',
    5: 'Paraná',
    6: 'Santa Catarina',
    7: 'Rio Grande do Sul',
    8: 'Distrito Federal',
    9: 'Ceará',
    10: 'Bahia',
    11: 'Rio Grande do Norte',
    12: 'Demais estados'
}
```

```python
novo_recode = {
    'Coluna': [],
    'Codigo': [],
    'Label': []
}

for codigo, label in BANDEIRA.items():
    novo_recode['Coluna'].append(NOME_BANDEIRA_RECODE)
    novo_recode['Codigo'].append(codigo)
    novo_recode['Label'].append(label)

lista_labels = pd.concat([lista_labels, pd.DataFrame(novo_recode)], axis=0, ignore_index=True)
display(lista_labels.loc[lista_labels['Coluna'] == NOME_BANDEIRA_RECODE])
```
